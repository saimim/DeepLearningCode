```python
from keras.layers import *
```

```python

# define two sets of inputs
inputA = Input(shape=(32,))
inputB = Input(shape=(128,))
```

```python

# the first branch operates on the first input
x = Dense(8, activation="relu")(inputA)
x1 = Dense(4, activation="relu")(x)
```

```python

y = Dense(64, activation="relu")(inputB)
y1 = Dense(32, activation="relu")(y)
y2 = Dense(4, activation="relu")(y1)
```

```python
combined = concatenate([x1, y2])
```

```python
z = Dense(2, activation="relu")(combined)
z1 = Dense(12, activation="relu")(z)
z2 = Dense(12, activation="relu")(z1)
```

```python
hidden1 = Dense(128,activation='relu')(z2)
hidden2 = Dense(64,activation='relu')(z2)
```

```python
hidden3 = Dense(128,activation='relu')(hidden1)
hidden4 = Dense(64,activation='relu')(hidden2)
```

```python
output1 = Dense(1,activation='linear')(hidden2)
output2 = Dense(1,activation='sigmoid')(hidden2)
```

```python
model = Model(inputs=[inputA, inputB],outputs=[output1,output2])
```
